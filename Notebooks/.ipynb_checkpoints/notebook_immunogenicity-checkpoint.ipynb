```python
### Inclusions ###

rootf = ... ## Your folder ##
# put here the folder where all subfolders (diffRBM, Align utils, Immunogenicity_model, Notebooks etc. are saved)

import sys, os, pickle

sys.path.append(rootf + '/Notebooks/')
sys.path.append(rootf + '/diffRBM/source/')
sys.path.append(rootf + '/diffRBM/utilities/')

import csv
import utilities as utilities
import rbm as rbm
import diffrbm as diffrbm

import dataset_utils, RBM_utils, evaluate_learning_utils
import utilities, sequence_logo,plots_utils
import Proteins_utils

import numpy as np
import pandas as pd
import importlib
import utilities_diffrbm

# Plots stuff
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import matplotlib as mpl
from matplotlib.backends.backend_pdf import PdfPages
from matplotlib import patches
from pandas.plotting import table
mpl.rcParams['font.family'] = ['Garuda']
mpl.rcParams['font.serif'] = ['Garuda-Oblique']

import random
import sklearn
from sklearn import metrics

import torch
import torch.nn
import torch.nn.functional as F
import torch.optim as optim
from scipy.optimize import fsolve
import math

## A series of useful functions ##

def overlap_seqs(list1,list2):
    overlap=[]
    for i in range(len(list1)):
        if list1[i] in list2:
            overlap.append(list1[i])
    return overlap

def align_seqs(all_cdr3,SA,SAmin,SAmax):
    import subprocess, os
    name_mat = rootf + '/Align_utils/align_protpy.py'
    all_seqs1 = list(all_cdr3)
    name_seed = rootf + '/Align_utils/prots_seed.txt'
    with open(name_seed, 'w') as out_f:
        for u in range(len(all_seqs1)):
            if len(all_seqs1[u]) >= SAmin:  
                out_f.write(all_seqs1[u] + '\n')
                
    subprocess.call('python3 ' + name_mat + ' -ss ' + name_seed + ' -SA ' + str(SA) + ' -SAmin ' + str(SAmin) + ' -SAmax ' + str(SAmax), shell = True)
    seqs_al = np.loadtxt(rootf + '/Align_utils/aligned_prot.txt')
    os.system('rm ' + rootf + '/Align_utils/aligned_prot.txt')
    
    return seqs_al

def align_seqs_seed(seed_seqs,new_seqs,SA,SAmin,SAmax):
    
    import subprocess, os
    name_mat = rootf + '/Align_utils/align_prot_to_seedpy.py'
    
    all_seqs1 = list(seed_seqs)
    name_seed = rootf + '/Align_utils/prots_seed.txt'
    with open(name_seed, 'w') as out_f:
        for u in range(len(all_seqs1)):
            if len(all_seqs1[u]) >= SAmin:  
                out_f.write(all_seqs1[u] + '\n')
                
    seqs_gapless = list(new_seqs)
    name_seqs = rootf + '/Align_utils/prots_seqs.txt'
    with open(name_seqs, 'w') as out_f:
        for u in range(len(seqs_gapless)):
            if len(seqs_gapless[u]) >= SAmin:  
                out_f.write(seqs_gapless[u] + '\n')
                
    subprocess.call('python3 ' + name_mat + ' -sseed ' + name_seed + ' -sseqs ' + name_seqs + ' -SA ' + str(SA) + ' -SAmin ' + str(SAmin) + ' -SAmax ' + str(SAmax) + ' -yw 0', shell = True)
    seqs_al = np.loadtxt(rootf + '/Align_utils/aligned_prot.txt')
    os.system('rm ' + rootf + '/Align_utils/aligned_prot.txt')
    
    return seqs_al

def flatten_list(listoflist):
    listoflist_fl = [];
    for l in range(len(listoflist)):
        for u in range(len(listoflist[l])):
            listoflist_fl.append(listoflist[l][u])
    return listoflist_fl
 
curr_int = np.int16
def convert_number(seqs): # convert to numbers already aligned seqs
    aa = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V',  'W', 'Y','-']
    aadict = {aa[k]: k for k in range(len(aa))} 
    
    msa_num = np.array(list(map(lambda x: [aadict[y] for y in x], seqs[0:])), dtype=curr_int, order="c") ### Here change ####
    
    return msa_num

def uniqueIndexes(l):
    seen = set()
    res = []
    for i, n in enumerate(l):
        if n not in seen:
            res.append(i)
            seen.add(n)
    return res

def convert_letter(seqs_n): # convert to numbers already aligned seqs
    aa = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V',  'W', 'Y','-']
    aadictinv = {k: aa[k] for k in range(len(aa))} 
    seqs=[]
    if type(seqs_n[0]) == curr_int:
        seqs.append(''.join([aadictinv[e] for e in seqs_n]))
    else:
        for t in range(len(seqs_n)):
            seqs.append(''.join([aadictinv[e] for e in seqs_n[t]]))
    return seqs

def produce_callback_weights(RBMdiff):
    def callback():
        RBMdiff.RBMpost.weights[RBMdiff.n_h_:] = 0
    return callback

def loglikelihood_indip_model(fields, logZ, seqs):
    return fields[np.arange(len(fields)), seqs].sum(axis=1) - logZ

def add_pseudocount(fields, n):
    return np.array([(f + 1/n) / np.sum((f + 1/n)) for f in fields])

def regularized_pwm(emp_freqsF, l2, pc):
    
    fields=[]
    pwmi=[]
    for ff in range(len(emp_freqsF)):
        def equations(x):
            return [(math.exp(x[i])/sum([math.exp(x[i]) for i in range(len(x))]) + l2*x[i] - emp_freqs[i]) for i in range(len(x))]
        emp_freqs = np.copy(emp_freqsF[ff])
        init = [np.log(c + pc) for c in emp_freqs]
        fields0 = fsolve(equations, init)
        pwmi0 = [math.exp(fields0[j])/sum([math.exp(fields0[i]) for i in range(len(fields0))]) for j in range(len(fields0))]
        
        fields.append(list(fields0))
        pwmi.append(pwmi0)
        
    fields=np.array(fields)
    pwmi=np.array(pwmi)
        
    return (fields, pwmi)

def regularized_diffpwm(emp_freqsF, l2, pc, fields_backF):
    
    fields=[]
    pwmi=[]
    for ff in range(len(emp_freqsF)):
        emp_freqs = np.copy(emp_freqsF[ff])
        fields_back = np.copy(fields_backF[ff])
        
        def equations(x):
            return [(math.exp(x[i] + fields_back[i])/sum([math.exp(x[i] + fields_back[i]) for i in range(len(x))]) + l2*(x[i]+ fields_back[i]) - emp_freqs[i]) for i in range(len(x))]
        
        
        init = [np.log(c + pc) for c in emp_freqs]
        fields0 = fsolve(equations, init)
        pwmi0 = [math.exp(fields0[j] + fields_back[j])/sum([math.exp(fields0[i] + fields_back[i]) for i in range(len(fields0))]) for j in range(len(fields0))]
        
        fields.append(list(fields0))
        pwmi.append(pwmi0)
        
    fields_top = np.array(fields)
    pwmi_top = np.array(pwmi)
        
    return (fields_top, pwmi_top)

import copy
def site_conservation_scores(RBM, seq, A=20, L=9):
    F = np.zeros((L, A))
    for i in range(L):
        for a in range(A):
            v = copy.copy(seq)
            v[i] = a
            F[i,a] = RBM.free_energy(v)
    return F.mean(1) - RBM.free_energy(v)
```

```python
## Consider Lethal mutation costs mutations along the peptide ##
import seaborn as sns

lw=3
binders_only = 0
also_weak = 0
no_pres = 1
colors=['r','b']

filename = name_folder + name_mf + '/fig3.xlsx'
file = pd.read_excel(filename, sheet_name= 'Fig. 3h', skiprows=3)
subs = list(file['Substitution'].values) + list(file['Substitution.1'].values) + list(file['Substitution.2'].values)
M_subs = list(file['M'].values) + list(file['M.1'].values) + list(file['M.2'].values)
file = pd.read_excel(filename, sheet_name= 'Fig. 3g', skiprows=3)
dd = (file['d_i'].values)[:9]

peps_names = ['NLVPMVATV','NLVPMVATV','NLVPMVATV', 'IMDQVPFSV', 'IMDQVPFSV', 'IMDQVPFSV', 'GRLKALCQR']
peps = ['peptide', 'peptide.1', 'peptide.2', 'peptide.3', 'peptide.4', 'peptide.5', 'peptide.6']

filename = name_folder + name_mf + '/suppl7.xlsx'
file = pd.read_excel(filename,sheet_name = 'ED Fig.  7c', skiprows=4)
ecs = ['obs_logC','obs_logC.1','obs_logC.2', 'obs_logC.3', 'obs_logC.4', 'obs_logC.5']

peps_n = convert_number(peps_names)

aa = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V',  'W', 'Y']

if also_weak:
    tcr_indices = [0,1,2,3,4,5]
else:
    tcr_indices = [0,1,2]
list_max=[]
mutants=[]
N=12
for n in range(N,N+1):
    mat_true_cost=[]
    mat_list=[]
    mat_list1=[]
    mat_list2=[]
    mat_list3=[]
    mat_nature_list=[]
    list_mut=[]
    list_pmt=[]
    for m in tcr_indices:

        pep = peps_names[m]
        codes = list(file[peps[m]].values)
        ecs_values = list(file[ecs[m]].values)

        mat_true = np.zeros((9,20))
        RBMq = dRBMf.top_rbm()
        g_top2 = RBMq.vlayer.fields

        l1_d = -dRBMf.RBMpost.free_energy(peps_n[m].astype(np.int16))
        l1_d = sum([RBM_back.vlayer.fields[i,peps_n[m][i]] for i in range(9)])
        l1 = -dRBMf.top_rbm().free_energy(peps_n[m].astype(np.int16))
        l1pwm = sum([(np.log(pwmP)-np.log(pwmB))[i,peps_n[m][i]] for i in range(9)])

        hm = RBMq.hlayer.mean_from_inputs(RBMq.vlayer.compute_output(peps_n[m], RBMq.weights, direction ='up') , beta = 1)
        I_top2 = RBMq.hlayer.compute_output(hm, RBMq.weights, direction='down')
        l10_3 = [g_top2[i, peps_n[m][i]] + sum(I_top2[:,i, peps_n[m][i]]) for i in range(9)]

        mat = np.zeros((9,20))
        mat1 = np.zeros((9,20))
        mat2 = np.zeros((9,20))
        mat3 = np.zeros((9,20))
        mat_nature = np.zeros((9,20))
        count=0
        count_n=0
        for i in range(9):
            for a in range(20):
                mat_nature[i,a] = dd[i]*M_data['M_ab'][code_s]
                code = pep[i] + str(i+1) + aa[a]

                if code in codes:
                    i_mt = codes.index(code)
                    p_mt = ecs_values[i_mt]
                    list_mut.append(code)
                    list_pmt.append(p_mt)
                    seq_new = np.copy(peps_n[m])
                    seq_new[i] = a
                    mutants.append(seq_new)
                    if p_mt:
                        count+=1
                        mat_true[i,a] = p_mt
                        list_mt.append(p_mt)

                        l2_d = sum([RBM_back.vlayer.fields[i,seq_new[i]] for i in range(9)])
                        l2 = -dRBMf.top_rbm().free_energy(seq_new.astype(np.int16))
                        
                        l2pwm = sum([(np.log(pwmP*pwmB))[i,seq_new[i]] for i in range(9)])
                        hm = RBMq.hlayer.mean_from_inputs(RBMq.vlayer.compute_output(seq_new, RBMq.weights, direction ='up') , beta = 1)
                        I_top2 = RBMq.hlayer.compute_output(hm, RBMq.weights, direction='down')
                        l10_3n = [g_top2[i, seq_new[i]] + sum(I_top2[:,i, seq_new[i]]) for i in range(9)]
                        mat[i,a] = l10_3n[i] - (l10_3[i])
                        mat1[i,a] = l10_3n[i] + l2_d - (l10_3[i]+l1_d)
                        mat2[i,a] = l2pwm - l1pwm
                        mat3[i,a] = l2 -l1
                        mutants.append(seq_new)

        mat_true_cost.append(mat_true) 
        mat_nature_list.append(mat_nature) 
        mat_list2.append(mat2)
        mat_list3.append(mat3)
        mat_list1.append(mat1)
        mat_list.append(mat)

    import scipy.stats

    ml = mat_list3
    label_method = 'Difference in top RBM scores'
    name_col = 'Difference top likelihood'

    list_m = []
    list_p = []
    listc=[]
    listc1=[]
    listc2=[]
    listc3=[]
    listc_nature=[]
    listpv=[]
    list_sc1=[]
    list_sc2=[]
    list_sc1b=[]
    list_sc2b=[]
    for u in range(len(mat_true_cost)):
        pp = np.array(mat_true_cost[u])
        list_p += list(pp[pp!=0])
        list_max.append(np.max(pp[pp!=0]))

        mm = np.array(ml[u])
        list_m += list(mm[pp!=0])
        listc.append(scipy.stats.spearmanr(pp[pp!=0],mm[pp!=0])[0])
        list_sc1 += list(pp[pp!=0])
        list_sc2 += list(mm[pp!=0])

        mm = np.array(mat_list1[u])
        listc1.append(scipy.stats.spearmanr(pp[pp!=0],mm[pp!=0])[0])

        mm = np.array(mat_list2[u])
        listc2.append(scipy.stats.spearmanr(pp[pp!=0],mm[pp!=0])[0])

        mm = np.array(mat_nature_list[u])

        listc_nature.append(scipy.stats.spearmanr(pp[pp!=0],mm[pp!=0])[0])

        mm = np.array(mat_list3[u])
        listc3.append(scipy.stats.spearmanr(pp[pp!=0],mm[pp!=0])[0])
        list_sc1b += list(pp[pp!=0])
        list_sc2b += list(mm[pp!=0])
        listpv.append(scipy.stats.spearmanr(pp[pp!=0],mm[pp!=0])[1])

list_sm0=[]
list_sm1=[]
list_sm2=[]
lsup=1
linf=10
for i in range(9):
    for a in range(20):
        if mat_true_cost[0][i,a]!=0 and mat_true_cost[0][i,a]==list_max[0]:
            list_sm0.append([i,a])
        if mat_true_cost[1][i,a]!=0 and mat_true_cost[1][i,a]==list_max[1]:
            list_sm1.append([i,a])
        if mat_true_cost[2][i,a]!=0 and mat_true_cost[2][i,a]==list_max[2]:
            list_sm2.append([i,a])

fin_sm=[]           
for i in range(9):
    for a in range(20):
        if [i,a] in list_sm0 and [i,a] in list_sm1 and [i,a] in list_sm2:
            fin_sm.append([i,aa[a]])

list_exclude=[]
if no_pres:
    list_exclude=[1,8] 

fin_wm=[]           
for i in range(9):
    for a in range(20):
        if [i,a] not in list_sm0 and [i,a] not in list_sm1 and [i,a] not in list_sm2 and i not in list_exclude:
            fin_wm.append([i,aa[a]])

list_high=[]
list_low=[]
h=0
for f in list_max[:3]:
    inds = [i for i in range(len(list_sc1)) if list_sc1[i]==f]
    list_high += [list_sc2[i] for i in inds]
    ind_pres=[]
    if no_pres:
        inds_pres = [n for n in np.arange(h*171,(h+1)*171) if list_sc2[n] in list(ml[h][1,:]) + list(ml[h][8,:])]
    list_high += [list_sc2[i] for i in inds]
    list_low += [list_sc2[i] for i in np.arange(h*171,(h+1)*171) if i not in inds and i not in inds_pres]
    h+=1

pointsPall=[]
pointsTall=[]
annotations=[]
pointsP=[]
pointsTs=[]
pointsTw=[]
mat_pos=np.zeros((3,9))
for h in [0,1,2]:
    for i in range(9):
        for a in range(20):
            if mat_true_cost[h][i,a]!=0 and [i,aa[a]] in fin_sm:
                annotations.append([i+1,aa[a]])
                pointsP.append(mat_true_cost[h][i,a])
                pointsTs.append(ml[h][i,a])
            if mat_true_cost[h][i,a]!=0 and [i,aa[a]] in fin_wm:

                annotations.append([i+1,aa[a]])
                pointsP.append(mat_true_cost[h][i,a])
                pointsTw.append(ml[h][i,a])

df = pd.read_csv(name_folder + name_mf + '/mutation_costs_control_all_positions.tsv', sep='\t')
llp = list_high
llc = list(df[name_col].values)
fig,ax=plt.subplots()
df1 = {'score': list(-np.array(llp))  + list(-np.array(llc)), 'name': list(np.repeat('Lethal',len(llp))) + list(np.repeat('Control',len(llc)))}
if save_files:
    df2 = pd.DataFrame(df1, columns = ['score','name'])
    df2.to_csv(name_folder + name_mf + '/high_cost.tsv', sep='\t', index=False)
df1 = pd.read_csv(name_folder + name_mf + '/high_cost.tsv', sep='\t')
sns.boxplot(y='score', x= 'name', data=df1, showfliers = False)
ax.set_ylabel(label_method, fontsize = s2)
ax.tick_params(axis='both', which='major', labelsize = s2)
from scipy import stats
print(stats.mannwhitneyu(llp,llc))
```
