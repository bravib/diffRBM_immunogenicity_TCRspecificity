Here I do the same that Barbara did for peptides. That is, I explore various hyperparamters (different number of HUs, different values of l1b, different values of l2f) and for each value I create 50 different pairs training-test, compute the performance on the test set, and in this way I fix the hyperparameters.

In this version (v3), I also check the new parameter for regularization (regularize only fields top, or the total fields back+top).

```python
import matplotlib.pyplot as plt
%matplotlib inline
import sys
sys.path.append('../../apps/diffRBM/source/')
sys.path.append('../../apps/diffRBM/utilities/')
from common_imports import set_num_threads
set_num_threads(2) # Set the number of cores. Must be executed before importing numpy&numba.
import numpy as np
import rbm, diffrbm, utilities, sequence_logo, Proteins_utils, RBM_utils
import pandas as pd
from scipy import stats
from sklearn import metrics

import itertools
```

```python
def shorten_gene_name(gene):
    """Removes 0 in gene name when not needed."""
    pre = gene[:4]
    first_num = str(int(gene[4:6]))
    if len(gene[6:]) == 0:
        short_gene = pre + first_num
    else:
        dash = gene[6:7]
        second_num = str(int(gene[7:]))
        short_gene = pre + first_num + dash + second_num
    return short_gene
```

```python
def add_VJ_info_2num(seqs_2num, V_list, J_list, V_dict, J_dict):
    Vlist_2num = np.array([V_dict[v] for v in V_list]).reshape((seqs_2num.shape[0], 1))
    Jlist_2num = np.array([J_dict[j] for j in J_list]).reshape((seqs_2num.shape[0], 1))
    seqs_2num_VJ = np.append(seqs_2num, Vlist_2num, axis=1)
    seqs_2num_VJ = np.append(seqs_2num_VJ, Jlist_2num, axis=1)
    return seqs_2num_VJ
```

```python
# load NLVPMVATV data, load JV2num_dict_train_train.csv, do num2seq plus add V and J

vdgdb_df = pd.read_csv('VDJdb_NLVPMVATV_WithAligned20.csv')
vdgdb_df = vdgdb_df.drop_duplicates().reset_index(drop=True)

# load JV2num dicts - see comment below
t_df = pd.read_csv('J2num_dict_train_train.csv')
J2num_dict = dict(zip(t_df['J_gene'], t_df['RBM_VU_id']))
t_df = pd.read_csv('V2num_dict_train_train.csv')
V2num_dict = dict(zip(t_df['V_gene'], t_df['RBM_VU_id']))

# do num2seq
t_allseq = vdgdb_df["ali_seq"].to_list()
allseq_2num = np.array(Proteins_utils.seq2num(t_allseq), dtype=np.int16)
# add to gen_seqs2num two integers for each row, obtained from V and J
Vlist = vdgdb_df["TRBV_gene"].to_list()
Jlist = vdgdb_df["TRBJ_gene"].to_list()

allseq_2num_withVJ = add_VJ_info_2num(allseq_2num, Vlist, Jlist, V2num_dict, J2num_dict)
```

```python
# load Emerson test set and prepare it for usage in ROC computations
emerson_test_set = pd.read_csv('train_test_WithoutDuplicates_aligned_20.csv.gz')
emerson_test_seqs2num =  np.array(Proteins_utils.seq2num(emerson_test_set['ali_seq'].to_list()), dtype=np.int16)
Vlist = emerson_test_set["v_gene"].to_list()
Jlist = emerson_test_set["j_gene"].to_list()
emerson_test_2num_withVJ = add_VJ_info_2num(emerson_test_seqs2num, Vlist, Jlist, V2num_dict, J2num_dict)
```

```python
# load backRBM
mybackRBM = RBM_utils.loadRBM('./backRBM-100hu-lowreg-withVJ-train_train_WithoutDuplicates.data')
```

# hyperparameter setting

Since this takes a while, I do the following: I saved a first run of vectors as temporary files, then I reload the vectors and I do not overwrite them anymore. After adding other parts, I update the temporary files. Notice that this only work if I have to block them explicitly, the files are not updated at each iteration.

```python
# load old files
try:
    ave_aveloglik_train = np.loadtxt('temp_ave_aveloglik_train.txt')
except:
    ave_aveloglik_train = []
try:
    std_aveloglik_train = np.loadtxt('temp_std_aveloglik_train.txt')
except:
    std_aveloglik_train = []
    
try:
    ave_aveloglik_test = np.loadtxt('temp_ave_aveloglik_test.txt')
except:
    ave_aveloglik_test = []
try:
    std_aveloglik_test = np.loadtxt('temp_std_aveloglik_test.txt')
except:
    std_aveloglik_test = []
    
try:
    ave_testset_ROC_post = np.loadtxt('temp_ave_testset_ROC_post.txt')
except:
    ave_testset_ROC_post = []
    
try:
    std_testset_ROC_post = np.loadtxt('temp_std_testset_ROC_post.txt')
except:
    std_testset_ROC_post = []
    
try:
    ave_testset_ROC_top = np.loadtxt('temp_ave_testset_ROC_top.txt')
except:
    ave_testset_ROC_top = []
    
try:
    std_testset_ROC_top = np.loadtxt('temp_std_testset_ROC_top.txt')
except:
    std_testset_ROC_top = []
```

```python
n_h_top_totest = [5, 10, 20, 40]
l1b_totest = [0.2, 0.1, 0.01]
Mt = len(vdgdb_df) - len(vdgdb_df) // 10 # size of training dataset of top
Mb = 1225743 # size of training dataset of back
l2f_totest = [0, 1/Mb, 1/Mt] # regularization of post fields (top+back)
l2ft_totest = [0, 1/Mt] # regularization of top units (top only)

N_times = 10

minlen_known = min(len(ave_aveloglik_train), len(ave_aveloglik_test), len(ave_testset_ROC_post),
                  len(std_testset_ROC_post), len(ave_testset_ROC_top), len(std_testset_ROC_top))

test_size = len(vdgdb_df) // 10 # in this way there are 454 for NLVPMVATV in the test set
c = 0
tested_hyperpars = []
for n_h, l1b, l2f, l2ft in itertools.product(n_h_top_totest, l1b_totest, l2f_totest, l2ft_totest):
    print("Number of HU tested:", n_h)
    print("Reg on weights (l1^2) tested:", l1b)
    print("Reg on post fields (l2) tested:", l2f)
    print("Reg on top fields (l2) tested:", l2ft)
    tested_hyperpars.append((n_h, l1b, l2f, l2ft))
    
    if c >= minlen_known:
        t_aveloglik_train = []
        t_aveloglik_test = []
        t_testset_ROC_top = []
        t_testset_ROC_post = []
        for _ in range(N_times):
            # prepare dataset
            np.random.shuffle(allseq_2num_withVJ)
            train_data = allseq_2num_withVJ[:-test_size]
            test_data = allseq_2num_withVJ[-test_size:]
            
            # train model
            batch_size = 400
            #n_updates = int(2e4)
            #n_iter = (n_updates // (train_data.shape[0] // batch_size))
            n_iter = 100
            learning_rate = None # default behaviour
            decay_after = 0.5
            N_MC = 10
            postRBM = rbm.RBM(n_h = mybackRBM.n_h + n_h, n_v = mybackRBM.n_v,
                      n_cv= mybackRBM.n_cv, visible='Potts', hidden='dReLU')
            diffRBM = diffrbm.DiffRBM(mybackRBM, postRBM)
            diffRBM.update_post_from_back(vlayer=True, hlayer=True)
            diffRBM.fit_top(train_data, batch_size = batch_size, n_iter = n_iter, l1b = l1b, l2_fields = l2f,
                            l2_fields_top = l2ft, N_MC = N_MC, decay_after = decay_after, verbose = 0, 
                            vverbose = 0, batch_norm=True)
    
            # compute loglikelihood (post)
            t_aveloglik_train.append(np.mean(diffRBM.RBMpost.likelihood(np.array(train_data, dtype=np.int16))))
            t_aveloglik_test.append(np.mean(diffRBM.RBMpost.likelihood(np.array(test_data, dtype=np.int16))))
            
            # compute performances (post and top)
            t_rbm = diffRBM.RBMpost
            liks_positive = -t_rbm.free_energy(np.array(test_data, dtype=np.int16))
            liks_negative = -t_rbm.free_energy(np.array(emerson_test_2num_withVJ, dtype=np.int16))
            labels = np.hstack((np.zeros((len(liks_negative))), np.ones((len(liks_positive))))) 
            scores = np.hstack((liks_negative, liks_positive))
            fpr, tpr, thresholds = metrics.roc_curve(labels, scores)
            t_testset_ROC_post.append(metrics.auc(fpr, tpr))
            
            t_rbm = diffRBM.top_rbm()
            liks_positive = -t_rbm.free_energy(np.array(test_data, dtype=np.int16))
            liks_negative = -t_rbm.free_energy(np.array(emerson_test_2num_withVJ, dtype=np.int16))
            labels = np.hstack((np.zeros((len(liks_negative))), np.ones((len(liks_positive))))) 
            scores = np.hstack((liks_negative, liks_positive))
            fpr, tpr, thresholds = metrics.roc_curve(labels, scores)
            t_testset_ROC_top.append(metrics.auc(fpr, tpr))
            
        # compute average and std dev
        ave_aveloglik_train = np.append(ave_aveloglik_train, np.mean(t_aveloglik_train))
        std_aveloglik_train = np.append(std_aveloglik_train, np.std(t_aveloglik_train))
        ave_aveloglik_test = np.append(ave_aveloglik_test, np.mean(t_aveloglik_test))
        std_aveloglik_test = np.append(std_aveloglik_test, np.std(t_aveloglik_test))
        ave_testset_ROC_top = np.append(ave_testset_ROC_top, np.mean(t_testset_ROC_top))
        std_testset_ROC_top = np.append(std_testset_ROC_top, np.std(t_testset_ROC_top))
        ave_testset_ROC_post = np.append(ave_testset_ROC_post, np.mean(t_testset_ROC_post))
        std_testset_ROC_post = np.append(std_testset_ROC_post, np.std(t_testset_ROC_post))
    
    print("Average loglik (post) for training set:", ave_aveloglik_train[c],
          " with standard devation:", std_aveloglik_train[c])
    print("Average loglik (post) for validation set:", ave_aveloglik_test[c],
          " with standard devation:", std_aveloglik_test[c])
    print("Average ROC post:", ave_testset_ROC_post[c], " with standard devation:", std_testset_ROC_post[c])
    print("Average ROC top:", ave_testset_ROC_top[c], " with standard devation:", std_testset_ROC_top[c])
    print("#########################################")
    c += 1
    # debug
    #if c == 1:
    #    break
```

# now let's train one for good
I will also increase to 5000 the number of epochs (Jerome's golden rule would be 2000), since for hyperparameter setting this would take too long but I hope that thanks to the regularization I cannot overfit even for longer trainings. However, I will check by looking at the final logliks and comparing them with those I obtained with the shorter training.
The reason why I use the longer trainin here is that Barbara showed that with longer training I obtain better results in the case with only fields (in this case the RBM should be equivalent to a PWM model, but it is such only when the trainig is longer).

```python
Mt = len(vdgdb_df) - len(vdgdb_df) // 10 # size of training dataset of top
Mb = 1225743 # size of training dataset of back
test_size = len(vdgdb_df) // 10 # in this way there are 454 for NLVPMVATV in the test set
np.random.shuffle(allseq_2num_withVJ)
train_data = allseq_2num_withVJ[:-test_size]
test_data = allseq_2num_withVJ[-test_size:]

# train model
n_h = 20
l1b = 0.01
l2f = 1/Mb
l2ft = 1/Mt
learning_rate = None # default behaviour
decay_after = 0.5
N_MC = 10
batch_size = 400
n_iter = 5000

postRBM = rbm.RBM(n_h = mybackRBM.n_h + n_h, n_v = mybackRBM.n_v,
                  n_cv= mybackRBM.n_cv, visible='Potts', hidden='dReLU')
diffRBM = diffrbm.DiffRBM(mybackRBM, postRBM)
diffRBM.update_post_from_back(vlayer=True, hlayer=True)
diffRBM.fit_top(train_data, batch_size = batch_size, n_iter = n_iter, l1b = l1b, l2_fields = l2f,
                l2_fields_top = l2ft, N_MC = N_MC, decay_after = decay_after, verbose = 0, 
                vverbose = 1, batch_norm=True)

# compute loglikelihood (post)
t_aveloglik_train = np.mean(diffRBM.RBMpost.likelihood(np.array(train_data, dtype=np.int16)))
t_aveloglik_test = np.mean(diffRBM.RBMpost.likelihood(np.array(test_data, dtype=np.int16)))

# compute performances (post and top)
t_rbm = diffRBM.RBMpost
liks_positive = -t_rbm.free_energy(np.array(test_data, dtype=np.int16))
liks_negative = -t_rbm.free_energy(np.array(emerson_test_2num_withVJ, dtype=np.int16))
labels = np.hstack((np.zeros((len(liks_negative))), np.ones((len(liks_positive))))) 
scores = np.hstack((liks_negative, liks_positive))
fpr, tpr, thresholds = metrics.roc_curve(labels, scores)
t_testset_ROC_post = metrics.auc(fpr, tpr)

t_rbm = diffRBM.top_rbm()
liks_positive = -t_rbm.free_energy(np.array(test_data, dtype=np.int16))
liks_negative = -t_rbm.free_energy(np.array(emerson_test_2num_withVJ, dtype=np.int16))
labels = np.hstack((np.zeros((len(liks_negative))), np.ones((len(liks_positive))))) 
scores = np.hstack((liks_negative, liks_positive))
fpr, tpr, thresholds = metrics.roc_curve(labels, scores)
t_testset_ROC_top = metrics.auc(fpr, tpr)
```

```python
# check performance - DIFF
t_rbm = diffRBM.RBMpost
liks_positive = -t_rbm.free_energy(np.array(test_data, dtype=np.int16))
liks_negative = -t_rbm.free_energy(np.array(emerson_test_2num_withVJ, dtype=np.int16))
labels = np.hstack((np.zeros((len(liks_negative))), np.ones((len(liks_positive))))) 
scores = np.hstack((liks_negative, liks_positive))
fpr, tpr, thresholds = metrics.roc_curve(labels, scores)
print("AUROC (DIFF):", metrics.auc(fpr, tpr))

plt.hist(liks_positive, density=True, histtype='step', fill=False, label="NLVPMVATV")
plt.hist(liks_negative, density=True, histtype='step',  fill=False, label="Emerson")
plt.legend()
```
